# Laporan Analisis Bab 8: Dimensionality Reduction

## Pendahuluan
[cite_start]Bab 8 membahas mengenai Reduksi Dimensi (Dimensionality Reduction), sebuah teknik penting dalam Machine Learning yang bertujuan untuk mengurangi jumlah fitur (dimensi) dalam dataset[cite: 260]. [cite_start]Masalah utama yang ingin diatasi oleh teknik ini adalah "kutukan dimensi" (curse of dimensionality), di mana kinerja model dapat menurun secara drastis pada dataset dengan dimensi yang sangat tinggi[cite: 260]. [cite_start]Meskipun reduksi dimensi dapat mempercepat pelatihan, penting untuk memahami bahwa ini juga dapat menyebabkan hilangnya informasi, yang berpotensi memengaruhi kinerja model secara keseluruhan[cite: 260].

## Ringkasan Isi Bab

Bab 8 menguraikan beberapa konsep kunci terkait Reduksi Dimensi:

1.  **The Curse of Dimensionality (Kutukan Dimensi)**:
    * [cite_start]Intuisi manusia terbatas pada tiga dimensi, membuat sulit membayangkan ruang berdimensi tinggi[cite: 261].
    * [cite_start]Banyak hal berperilaku sangat berbeda di ruang berdimensi tinggi; misalnya, sebagian besar titik di *hypercube* berdimensi tinggi sangat dekat dengan batas[cite: 261].
    * [cite_start]Dataset berdimensi tinggi cenderung sangat jarang (sparse), dengan instance pelatihan yang jauh satu sama lain[cite: 262].
    * [cite_start]Prediksi menjadi kurang dapat diandalkan karena ekstrapolasi yang lebih besar diperlukan[cite: 262].
    * [cite_start]Semakin banyak dimensi, semakin tinggi risiko *overfitting*[cite: 262].
    * [cite_start]Jumlah instance pelatihan yang diperlukan untuk mencapai kepadatan tertentu tumbuh secara eksponensial dengan jumlah dimensi[cite: 262].

2.  **Main Approaches for Dimensionality Reduction (Pendekatan Utama untuk Reduksi Dimensi)**:
    * **Projection (Proyeksi)**:
        * [cite_start]Umumnya, instance pelatihan tidak tersebar merata di semua dimensi[cite: 262]. [cite_start]Banyak fitur hampir konstan atau sangat berkorelasi[cite: 262].
        * [cite_start]Instance pelatihan seringkali terletak dekat dengan subruang berdimensi rendah dari ruang berdimensi tinggi[cite: 262].
        * [cite_start]Proyeksi melibatkan pemetaan data ke subruang berdimensi lebih rendah ini[cite: 263].
    * **Manifold Learning**:
        * [cite_start]Konsep *manifold* menggambarkan bentuk berdimensi rendah yang dapat ditekuk dan dipelintir dalam ruang berdimensi lebih tinggi (misalnya, Swiss roll adalah *manifold* 2D dalam ruang 3D)[cite: 265].
        * [cite_start]Algoritma *Manifold Learning* bekerja dengan memodelkan *manifold* tempat instance pelatihan berada[cite: 265].
        * [cite_start]Ini didasarkan pada asumsi *manifold* (atau hipotesis *manifold*), yang menyatakan bahwa sebagian besar dataset berdimensi tinggi di dunia nyata terletak dekat dengan *manifold* berdimensi jauh lebih rendah[cite: 265].
        * [cite_start]Asumsi implisit sering menyertainya: tugas (misalnya, klasifikasi) akan lebih sederhana jika diekspresikan dalam ruang berdimensi rendah dari *manifold*[cite: 265]. [cite_start]Namun, asumsi ini tidak selalu berlaku[cite: 265].

3.  **PCA (Principal Component Analysis)**:
    * [cite_start]Algoritma reduksi dimensi yang paling populer[cite: 266].
    * [cite_start]**Preserving the Variance (Mempertahankan Varians)**: Memilih *hyperplane* yang paling dekat dengan data, yaitu yang mempertahankan jumlah varians maksimum[cite: 266]. [cite_start]Ini juga meminimalkan jarak kuadrat rata-rata antara dataset asli dan proyeksinya[cite: 267].
    * [cite_start]**Principal Components (Komponen Utama)**: PCA mengidentifikasi sumbu (komponen utama) yang menyumbang jumlah varians terbesar dalam dataset[cite: 267]. [cite_start]Sumbu kedua orthogonal terhadap yang pertama dan menyumbang varians terbesar yang tersisa, dan seterusnya[cite: 267]. [cite_start]Vektor unit yang mendefinisikan komponen utama (PC) dapat ditemukan menggunakan Singular Value Decomposition (SVD)[cite: 268].
    * [cite_start]**Projecting Down to d Dimensions (Memproyeksikan ke d Dimensi)**: Dataset dapat direduksi ke $d$ dimensi dengan memproyeksikannya ke *hyperplane* yang ditentukan oleh $d$ komponen utama pertama[cite: 268].
    * [cite_start]**Using Scikit-Learn**: Kelas `PCA` di Scikit-Learn menggunakan dekomposisi SVD dan secara otomatis menangani pemusatan data[cite: 269].
    * [cite_start]**Explained Variance Ratio**: Variabel `explained_variance_ratio_` menunjukkan proporsi varians dataset yang terletak di sepanjang setiap komponen utama[cite: 269]. [cite_start]Ini membantu dalam memilih jumlah dimensi yang tepat untuk dipertahankan (misalnya, 95% varians)[cite: 270].
    * [cite_start]**PCA for Compression (PCA untuk Kompresi)**: Setelah reduksi dimensi, dataset membutuhkan lebih sedikit ruang[cite: 271]. [cite_start]Dataset yang direduksi dapat didekompresi kembali ke dimensi asli (dengan sedikit kehilangan informasi)[cite: 271]. [cite_start]Kesalahan rekonstruksi (mean squared distance antara data asli dan data yang direkonstruksi) dapat diukur[cite: 271].
    * [cite_start]**Randomized PCA**: Menggunakan algoritma stokastik untuk menemukan perkiraan komponen utama pertama dengan cepat, lebih cepat daripada SVD penuh ketika $d$ jauh lebih kecil dari $n$[cite: 272].
    * [cite_start]**Incremental PCA (IPCA)**: Memungkinkan reduksi dimensi pada dataset besar yang tidak muat dalam memori utama, dengan memproses data dalam *mini-batches*[cite: 272].

4.  **Kernel PCA (kPCA)**:
    * [cite_start]Menerapkan "kernel trick" (seperti pada SVM) untuk melakukan proyeksi non-linear yang kompleks untuk reduksi dimensi[cite: 274].
    * [cite_start]Seringkali baik dalam mempertahankan kluster instance setelah proyeksi atau bahkan "menggulirkan" dataset yang dekat dengan *manifold* yang terpuntir[cite: 274].
    * [cite_start]**Selecting a Kernel and Tuning Hyperparameters (Memilih Kernel dan Menyetel Hyperparameter)**: Karena ini adalah algoritma pembelajaran tanpa pengawasan, tidak ada ukuran kinerja yang jelas[cite: 274]. [cite_start]Umumnya, digunakan sebagai langkah persiapan untuk tugas pembelajaran terawasi, dan *grid search* dapat digunakan untuk memilih kernel dan hyperparameter yang memberikan kinerja terbaik pada tugas tersebut[cite: 275]. [cite_start]Pendekatan lain yang tanpa pengawasan adalah meminimalkan kesalahan rekonstruksi pra-citra[cite: 276].

5.  **LLE (Locally Linear Embedding)**:
    * [cite_start]Teknik reduksi dimensi non-linear (NLDR) yang kuat lainnya[cite: 277].
    * [cite_start]Ini adalah teknik *Manifold Learning* yang tidak bergantung pada proyeksi[cite: 277].
    * [cite_start]Bekerja dengan mengukur bagaimana setiap instance pelatihan secara linier berhubungan dengan tetangga terdekatnya, dan kemudian mencari representasi berdimensi rendah dari dataset di mana hubungan lokal ini paling dipertahankan[cite: 278].
    * [cite_start]Sangat baik dalam membuka *manifold* yang terpuntir, terutama dengan sedikit gangguan (*noise*)[cite: 277].

6.  **Other Dimensionality Reduction Techniques (Teknik Reduksi Dimensi Lainnya)**:
    * [cite_start]**Random Projections**: Memproyeksikan data ke ruang berdimensi lebih rendah menggunakan proyeksi linier acak yang cenderung mempertahankan jarak dengan baik[cite: 282].
    * [cite_start]**Multidimensional Scaling (MDS)**: Mengurangi dimensi sambil mencoba mempertahankan jarak antara instance[cite: 282].
    * [cite_start]**Isomap**: Membuat grafik dengan menghubungkan setiap instance ke tetangga terdekatnya, lalu mengurangi dimensi sambil mencoba mempertahankan jarak geodetik antara instance[cite: 283].
    * [cite_start]**t-Distributed Stochastic Neighbor Embedding (t-SNE)**: Mengurangi dimensi sambil mencoba menjaga instance serupa tetap dekat dan instance yang tidak serupa tetap terpisah[cite: 280]. [cite_start]Ini terutama digunakan untuk visualisasi kluster instance di ruang berdimensi tinggi[cite: 280].
    * [cite_start]**Linear Discriminant Analysis (LDA)**: Meskipun merupakan algoritma klasifikasi, selama pelatihan, ia mempelajari sumbu paling diskriminatif antara kelas-kelas, yang kemudian dapat digunakan untuk memproyeksikan data[cite: 280]. [cite_start]Bermanfaat untuk reduksi dimensi sebelum algoritma klasifikasi lain[cite: 280].

```python

# Import library yang diperlukan
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll, load_sample_image, load_digits
from sklearn.decomposition import PCA, KernelPCA, IncrementalPCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.manifold import LocallyLinearEmbedding, TSNE # Untuk t-SNE

# Mengatur random state untuk reproducibility
np.random.seed(42)

# --- BAGIAN 1: PCA (PRINCIPAL COMPONENT ANALYSIS) ---
print("--- Principal Component Analysis (PCA) ---")

# 1.1 Menghasilkan data linear-looking untuk demonstrasi SVD dan proyeksi
print("\n1.1 Demonstrasi SVD dan Proyeksi Manual:")
X_linear = 2 * np.random.rand(100, 1) # Asumsi data 1D asli
y_linear = 4 + 3 * X_linear + np.random.randn(100, 1)

# Menggabungkan X_linear dan y_linear menjadi data 2D untuk PCA
# Misal, data yang tersebar di sekitar garis y=3x+4
# Kita buat data yang lebih mirip contoh buku (seperti di Bab 4)
# X_2D = np.random.rand(100, 2)
# X_2D_centered = X_2D - X_2D.mean(axis=0)

# Untuk PCA, kita butuh data dengan varians tersebar.
# Mari buat data 3D yang cenderung linear seperti contoh di buku (Gambar 8-2)
# Data yang hampir linear di 3D, tapi aslinya 'hanya' 2D
X_3D = 0.5 * np.random.rand(200, 3) # Beberapa random data
X_3D[:, 0] = X_3D[:, 0] * 2 + 1 # Rentang yang lebih luas untuk satu dimensi
X_3D[:, 1] = X_3D[:, 1] * 3 + 2 # Rentang yang lebih luas untuk dimensi lain
X_3D[:, 2] = X_3D[:, 2] * 0.1 # Dimensi ketiga dengan varians kecil
# Tambahkan noise agar tidak sempurna linear
X_3D += np.random.randn(200, 3) * 0.1

X_centered = X_3D - X_3D.mean(axis=0)

# Menghitung SVD
U, s, Vt = np.linalg.svd(X_centered)

# Mendapatkan dua komponen utama pertama
c1 = Vt.T[:, 0]
c2 = Vt.T[:, 1]
print(f"Vektor Komponen Utama 1 (c1): {c1}")
print(f"Vektor Komponen Utama 2 (c2): {c2}")

# Memproyeksikan data ke 2 dimensi
W2 = Vt.T[:, :2] # Mengambil 2 komponen utama pertama
X2D = X_centered.dot(W2)
print(f"Bentuk data setelah proyeksi ke 2D: {X2D.shape}")

# 1.2 Menggunakan PCA dari Scikit-Learn
print("\n1.2 Menggunakan PCA dari Scikit-Learn:")
pca = PCA(n_components=2)
X2D_sklearn = pca.fit_transform(X_3D) # PCA secara otomatis melakukan centering
print(f"Bentuk data setelah PCA Scikit-Learn ke 2D: {X2D_sklearn.shape}")
print(f"Komponen Utama dari Scikit-Learn PCA (transposed): \n{pca.components_}")

# 1.3 Explained Variance Ratio
print("\n1.3 Explained Variance Ratio:")
# Latih PCA tanpa mengurangi dimensi untuk mendapatkan semua explained variance ratio
pca_full = PCA()
pca_full.fit(X_3D)
explained_variance_ratios = pca_full.explained_variance_ratio_
print(f"Explained Variance Ratios: {explained_variance_ratios}")

# Memilih jumlah dimensi yang tepat (misal, untuk 95% varians)
cumsum = np.cumsum(explained_variance_ratios)
d = np.argmax(cumsum >= 0.95) + 1
print(f"Jumlah dimensi yang diperlukan untuk mempertahankan 95% varians: {d}")

# Latih ulang PCA dengan jumlah komponen yang dipilih
pca_95 = PCA(n_components=0.95) # Secara otomatis memilih jumlah komponen untuk 95% varians
X_reduced_95 = pca_95.fit_transform(X_3D)
print(f"Bentuk data setelah PCA 95% varians: {X_reduced_95.shape}")

# 1.4 PCA untuk Kompresi (menggunakan dataset MNIST)
print("\n1.4 PCA untuk Kompresi (Dataset MNIST):")
# Memuat dataset MNIST (perlu scikit-learn versi terbaru atau fetch_openml)
digits = load_digits()
X_mnist = digits.data
y_mnist = digits.target

# Memisahkan data menjadi training dan testing set
X_train_mnist, X_test_mnist, y_train_mnist, y_test_mnist = train_test_split(X_mnist, y_mnist, random_state=42)

# Mengurangi dimensi MNIST sambil mempertahankan 95% varians
pca_mnist = PCA(n_components=0.95)
X_reduced_mnist = pca_mnist.fit_transform(X_train_mnist)
print(f"Dimensi asli MNIST: {X_train_mnist.shape[1]}")
print(f"Dimensi setelah reduksi (95% varians): {X_reduced_mnist.shape[1]}")

# Rekonstruksi data
X_recovered_mnist = pca_mnist.inverse_transform(X_reduced_mnist)
print(f"Bentuk data setelah rekonstruksi: {X_recovered_mnist.shape}")

# Tampilkan contoh gambar asli dan yang direkonstruksi (untuk visualisasi)
def plot_digits(instances, images_per_row=5, **options):
    # The MNIST images are 8x8
    img_size = 8
    images_per_row = min(len(instances), images_per_row)
    images = [instance.reshape(img_size, img_size) for instance in instances]
    n_rows = (len(instances) - 1) // images_per_row + 1
    row_images = []
    for row in range(n_rows):
        r_images = images[row * images_per_row : (row + 1) * images_per_row]
        # Pad the last row with blank images if necessary
        if len(r_images) < images_per_row:
            blank_images = [np.zeros((img_size, img_size))] * (images_per_row - len(r_images))
            r_images.extend(blank_images)
        row_images.append(np.concatenate(r_images, axis=1))
    image = np.concatenate(row_images, axis=0)
    plt.imshow(image, cmap = plt.cm.binary, **options)
    plt.axis("off")

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plot_digits(X_train_mnist[::200], images_per_row=5) # Beberapa gambar asli
plt.title("Gambar Asli")
plt.subplot(1, 2, 2)
plot_digits(X_recovered_mnist[::200], images_per_row=5) # Gambar yang direkonstruksi
plt.title("Gambar Direkonstruksi (95% varians)")
plt.show()

# 1.5 Randomized PCA
print("\n1.5 Randomized PCA:")
# Lebih cepat untuk d jauh lebih kecil dari n
rnd_pca = PCA(n_components=X_reduced_mnist.shape[1], svd_solver="randomized", random_state=42)
X_reduced_rnd = rnd_pca.fit_transform(X_train_mnist)
print(f"Bentuk data setelah Randomized PCA: {X_reduced_rnd.shape}")

# 1.6 Incremental PCA (IPCA)
print("\n1.6 Incremental PCA (IPCA):")
# Untuk dataset yang tidak muat di memori
n_batches = 10 # Reduced from 100 to 10 to increase batch size
inc_pca = IncrementalPCA(n_components=X_reduced_mnist.shape[1])
for X_batch in np.array_split(X_train_mnist, n_batches):
    inc_pca.partial_fit(X_batch) # Melatih secara inkremental

X_reduced_inc = inc_pca.transform(X_train_mnist)
print(f"Bentuk data setelah Incremental PCA: {X_reduced_inc.shape}")
```

```python
# --- BAGIAN 2: KERNEL PCA (kPCA) ---
print("\n--- Kernel PCA (kPCA) ---")

# 2.1 Menggunakan kPCA pada Swiss Roll Dataset
print("\n2.1 Menggunakan kPCA pada Swiss Roll Dataset:")
# Menghasilkan data Swiss Roll (memerlukan scikit-learn >= 0.24)
try:
    X_swiss, t_swiss = make_swiss_roll(n_samples=1000, noise=0.2, random_state=42)

    # Menggunakan Kernel PCA dengan RBF kernel
    rbf_pca = KernelPCA(n_components=2, kernel="rbf", gamma=0.04)
    X_reduced_kpca = rbf_pca.fit_transform(X_swiss)

    print(f"Bentuk data Swiss Roll setelah kPCA: {X_reduced_kpca.shape}")

    # Tampilkan hasil kPCA (scatter plot)
    plt.figure(figsize=(8, 6))
    plt.scatter(X_reduced_kpca[:, 0], X_reduced_kpca[:, 1], c=t_swiss, cmap=plt.cm.hot)
    plt.title("Swiss Roll Dataset setelah kPCA (RBF Kernel)")
    plt.xlabel("Komponen Utama 1")
    plt.ylabel("Komponen Utama 2")
    plt.colorbar(label="Jarak Sepanjang Roll")
    plt.grid(True)
    plt.show()

except ImportError:
    print("Scikit-learn versi 0.24 atau lebih tinggi diperlukan untuk make_swiss_roll dengan noise.")

# 2.2 Memilih Hyperparameter kPCA menggunakan Grid Search dan Logistik Regression (Unsupervised task evaluation)
print("\n2.2 Memilih Hyperparameter kPCA dengan Grid Search:")
# Menggunakan dataset MNIST yang sudah dimuat
X_train_kpca, X_test_kpca, y_train_kpca, y_test_kpca = train_test_split(X_mnist, y_mnist, random_state=42)

# Membuat pipeline: kPCA -> Logistic Regression
# Mengurangi dimensi ke 50 (nilai arbitrer untuk contoh)
# Menggunakan dataset kecil untuk demonstrasi agar lebih cepat
X_small = X_train_kpca[:1000]
y_small = y_train_kpca[:1000]

clf = Pipeline([
    ("kpca", KernelPCA(n_components=50)),
    ("log_reg", LogisticRegression(solver="liblinear", random_state=42))
])

param_grid = [{
    "kpca__gamma": np.linspace(0.03, 0.05, 10),
    "kpca__kernel": ["rbf"]
}]

grid_search = GridSearchCV(clf, param_grid, cv=3, scoring="accuracy", n_jobs=-1)
# grid_search.fit(X_train_kpca, y_train_kpca) # Gunakan dataset penuh jika memungkinkan
grid_search.fit(X_small, y_small) # Gunakan dataset kecil untuk demonstrasi

print(f"Best parameters found by grid search: {grid_search.best_params_}")
print(f"Best cross-validation accuracy: {grid_search.best_score_:.4f}")

# 2.3 Meminimalkan Reconstruction Pre-image Error (Unsupervised task evaluation)
print("\n2.3 Meminimalkan Reconstruction Pre-image Error:")
# Menggunakan dataset kecil dari MNIST
# Reconstruct data (hanya bekerja jika kPCA dilatih dengan fit_inverse_transform=True)
# kPCA dengan fit_inverse_transform=True
rbf_pca_recon = KernelPCA(n_components=50, kernel="rbf", gamma=0.04,
                          fit_inverse_transform=True) # Memungkinkan inverse_transform

# Latih kPCA pada data kecil
rbf_pca_recon.fit(X_small)

# Transform dan inverse transform
X_reduced_small = rbf_pca_recon.transform(X_small)
X_preimage_small = rbf_pca_recon.inverse_transform(X_reduced_small)

# Hitung Mean Squared Error (MSE) antara data asli dan pre-image
mse = np.mean(np.sum(np.square(X_small - X_preimage_small), axis=1))
print(f"Mean Squared Error (MSE) Reconstruction Pre-image: {mse:.4f}")

# Untuk menemukan gamma yang terbaik, kita bisa menggunakan Grid Search lagi,
# tapi kali ini mencari gamma yang meminimalkan MSE rekonstruksi.
# Namun, inverse_transform tidak tersedia untuk semua kernel dan parameter.
# Scikit-learn >= 0.20 punya 'reconstruction_error' method for KernelPCA
# yang lebih handal.

# Cara alternatif (jika fit_inverse_transform=True tidak didukung atau tidak bekerja):
# Latih kPCA untuk setiap gamma, lalu hitung MSE pada data yang direkonstruksi.
# Ini bisa lambat. Grid Search dengan pipeline LogReg lebih umum untuk memilih kPCA.
```
